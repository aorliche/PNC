```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

csv = '/home/anton/Documents/Tulane/Research/Work/Datamining/Tutorial/diabetes_data_upload.csv'
df = pd.read_csv(csv)

print('Complete')
```

```python
df
```

```python
rcparams = {'figure.dpi': 80}
sns.set(rcparams)
sns.barplot(data=df, x='Polyuria', y='Age', hue='Gender', palette='husl')
plt.title('Polyuria Analysis')
plt.legend(bbox_to_anchor=(1.02,1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
sns.violinplot(data=df, x='delayed healing', y='Age', hue='Gender', palette='light:#5A9')
plt.title('Delayed Healing')
plt.show()
```
