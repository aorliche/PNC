```python
# Read the weirdly formatted csv file
# Weirdly formatted in that it's maybe less suitable for pandas that 

import re
import math

fields = ['Age', 
          'Sex', 
          'Height (cm)', 
          'Weight (kg)', 
          'Ethnicity', 
          'Race', 
          'Subject Education', 
          'Subject Socioeconomic Status',
          'Dad Education', 
          'Mom Education', 
          'Parental Socioeconomic status',
          'Neuroleptic Naive', 
          'Illness Duration']

field_num = ['03', '04', '06', '07', '08', '09', '10', '11', '13', '14', '15', '16', '17']
num_fields = ['Age', 'Height (cm)', 'Weight (kg)', 'Subject Education', 
              'Dad Education', 'Mom Education', 'Illness Duration']

num2field = dict()
demo = dict()

for f,n in zip(fields, field_num):
    num2field[n] = f
    f = re.sub(' ', '_', f)
    demo[f] = dict()

csv = open('/home/anton/Documents/Tulane/Research/MCIC/schizconnect_MCICShare_assessmentData_17461.csv', 'r')

for line in csv:
    if not re.search('>BIF', line):
        continue
    m = re.search(',(A[0-9]+),.*MCICSHARE_(\d\d),(.*)$', line)
    sub = m.group(1)
    num = m.group(2)
    val = m.group(3)
    if num2field[num] in num_fields:
        try:
            val = float(val)
            if math.isnan(val):
                continue
        except:
            continue
    else:
        val = val.strip()
#     print(sub,num,val)
    f = re.sub(' ', '_', num2field[num])
    demo[f][sub] = val
    
for f in demo:
    print(len(demo[f]))
```

```python
import pickle

imagedir = '/home/anton/Documents/Tulane/Research/ImageNomer/data/anton/cohorts/MCIC'

pickle.dump(demo, open(f'{imagedir}/demographics.pkl', 'wb'))

print('Complete')
```
