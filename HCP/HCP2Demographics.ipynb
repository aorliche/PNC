```python
# Create demographics file based on select field in two .csv files
# Sanity test regression of age, sex, race on REST FCs

import pandas as pd

dfRestrict = pd.read_csv('/home/anton/Documents/Yale/HCP/RESTRICTED_sony5650_6_18_2017_10_24_27.csv')

dfRestrict
```

```python
dfUnrestrict = pd.read_csv('/home/anton/Documents/Yale/HCP/unrestricted_aorlichenko_3_17_2025_9_18_27.csv')

dfUnrestrict
```

```python
# Removed: 8 HbA1C

restrictFields = '''Age_in_Yrs
Race
Weight
Height
BMI
Hematocrit_1
BPDiastolic
BPSystolic
Total_Drinks_7days
Total_Any_Tobacco_7days
SSAGA_Mj_Use'''

restrictFields = restrictFields.split()
restrictFields
```

```python
unrestrictFields = '''Gender
PSQI_Score
Strength_Unadj'''

unrestrictFields = unrestrictFields.split()
unrestrictFields
```

```python
import math

demo = {}

nansCount = 0

for field in restrictFields:
    demo[field] = {}

for field in unrestrictFields:
    demo[field] = {}

nansubs = 0

for index, row in dfRestrict.iterrows():
    sub = row['Subject']
    hasnan = False
    for field in restrictFields:
        val = row[field]
        if type(val) != str and math.isnan(val):
            nansCount += 1
            hasnan = True
            continue
        demo[field][sub] = val
    if hasnan == True:
        nansubs += 1

print('Done')
print(nansCount)
print(nansubs)
```

```python
nansCount2 = 0

for index, row in dfUnrestrict.iterrows():
    sub = row['Subject']
    for field in unrestrictFields:
        val = row[field]
        if type(val) != str and math.isnan(val):
            nansCount2 += 1
            continue
        demo[field][sub] = val

print('Done')
print(nansCount2)
```

```python
# Dump demographics

import pickle

pickle.dump(demo, open('/home/anton/Documents/Yale/HCP/demographics.pkl', 'wb'))

print('Done')
```

```python
# Get subs with all values

allFields = restrictFields + unrestrictFields
keepSubs = []

for sub in demo['Age_in_Yrs']:
    exclude = False
    for field in allFields:
        if sub not in demo[field]:
            exclude = True
    if not exclude:
        keepSubs.append(sub)

print(len(keepSubs))
```

```python
# Get FC

import numpy as np

fcsPkl = pickle.load(open('/home/anton/Documents/Yale/HCP/fcs_ya.pkl', 'rb'))
fcs = []

for sub in keepSubs:
    sub = str(sub)
    if sub in fcsPkl:
        fcs.append(fcsPkl[sub])

x = np.stack(fcs)

print(len(x))
```

```python
# Get target variable

field = 'Gender'
y = []

for sub in keepSubs:
    if str(sub) in fcsPkl:
        y.append(int(demo[field][sub] == 'F'))

y = np.array(y)

print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, LogisticRegression

def rmse(a, b):
    return np.mean((a-b)**2)**0.5

xtr, xt, ytr, yt = train_test_split(x, y, stratify=y, train_size=0.8)

reg = LogisticRegression(C=100).fit(xtr, ytr)
yhat = reg.predict(xt)
print(np.mean(yt == yhat))
print(np.mean(ytr))
# print(rmse(yt, yhat))
# print(rmse(yt, np.mean(ytr)))
```
