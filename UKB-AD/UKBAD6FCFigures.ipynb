```python
import numpy as np
from pathlib import Path

fcs = []

for ent in Path('/home/anton/Documents/Tulane/Research/ImageNomer/data/UKB3/conn/').iterdir():
    if '20227-2' in str(ent):
        p = np.load(str(ent))
        fcs.append(p)
    if len(fcs) >= 10:
        break

print('Done')
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm

def vec2mat(v):
    a,b = np.triu_indices(264,1)
    m = np.zeros((264,264))
    m[a,b] = v
    return m+m.T

plt.imshow(vec2mat(fcs[5]))
plt.xticks([])
plt.yticks([])
plt.colorbar()
plt.show()
```
