```python
import pickle

# Load meta dict

with open('../../PNC/MegaMeta.bin', 'rb') as f:
    meta = pickle.load(f)
    
print(len(list(meta.keys())))
```

```python
count = 0
countr = 0
countn = 0
counte = 0

for key in meta:
    sub = meta[key]
    if (('emoid' in sub and 'badEmoid' not in sub) 
        or ('nback' in sub and 'badNback' not in sub) 
        or ('rest' in sub and 'badRest' not in sub)):
        count += 1
    if ('rest' in sub and 'badRest' not in sub):
        countr += 1
    if ('nback' in sub and 'badNback' not in sub):
        countn += 1
    if ('emoid' in sub and 'badEmoid' not in sub):
        counte += 1

print(count)
print(countr)
print(countn)
print(counte)
```

```python
import csv

pncDir = '../../PNC_Good'

# Read wrat

wratFile = f'{pncDir}/wrat2.csv'
wratRaw = dict()
wratStd = dict()

with open(wratFile, newline='') as f:
    reader = csv.reader(f, delimiter=',')
    first = True
    for row in reader:
        if first:
            first = False
            continue
        try:
            key = int(row[0])
            wratRaw[key] = int(row[2])
            wratStd[key] = int(row[3])
        except Exception:
            print(f'Alert {row}')
```

```python
for key in meta:
    sub = meta[key]
    if (('emoid' in sub and 'badEmoid' not in sub) 
        or ('nback' in sub and 'badNback' not in sub) 
        or ('rest' in sub and 'badRest' not in sub)):
        if 'wrat' not in sub['meta']:
            print(f"Don't have WRAT for {key}")
            if key in wratRaw:
                print('Have in wrat2')
        else:
            if int(sub['meta']['wrat']['raw']) != wratRaw[key]:
                print(f"Different wrats {sub['meta']['wrat']['raw']} {wratRaw[key]}")
            if int(sub['meta']['wrat']['std']) != wratStd[key]:
                print(f"Different wrats {sub['meta']['wrat']['std']} {wratStd[key]}")
    if ('nback' in sub and 'badNback' not in sub):
        countn += 1
    if ('emoid' in sub and 'badEmoid' not in sub):
        counte += 1
```

```python
meta3 = dict()

for key in meta:
    sub = meta[key]
    if (('emoid' in sub and 'badEmoid' not in sub) 
        or ('nback' in sub and 'badNback' not in sub) 
        or ('rest' in sub and 'badRest' not in sub)):
        d = meta3[key] = dict()
        if 'wrat' in sub['meta']:
            d['wratRaw'] = wratRaw[key]
            d['wratStd'] = wratStd[key]
        d['AgeInMonths'] = int(sub['meta']['AgeInMonths'])
        d['Gender'] = sub['meta']['Gender']
        d['Ethnicity'] = sub['meta']['Ethnicity']
        if key not in meta2:
            print(f'meta2 does not have {key}')
        else:
            if d['AgeInMonths'] != meta2[key]['AgeInMonths']:
                print(f'Alert age {key}')
            if d['Gender'] != meta2[key]['Gender']:
                print(f'Alert gender {key}')
            if d['Ethnicity'] != meta2[key]['Ethnicity']:
                print(f'Alert ethnicity {key}')
    if ('rest' in sub and 'badRest' not in sub):
        d['rest'] = sub['restData'].astype('float32')
    if ('nback' in sub and 'badNback' not in sub):
        d['nback'] = sub['nbackData'].astype('float32')
    if ('emoid' in sub and 'badEmoid' not in sub):
        d['emoid'] = sub['emoidData'].astype('float32')
        
print('Complete')
print(len(list(meta3.keys())))
```

```python
# Save meta3 to file

import pickle

outputFile = f'{pncDir}/MegaMeta3.pkl'

with open(outputFile, 'wb') as f:
    pickle.dump(meta3, f)

print('Complete')
```
