```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,3,figsize=(16,5))

fields = 169

cats = ['FC', 'FC\n(SUB use)', 'AE Z', 'AE Z\n(SUB use)', 'VAE Z', 'VAE Z\n(SUB use)']
rest_counts = [71, 1, 85, 2, 13, 2]
nback_counts = [91, 0, 92, 1, 13, 0]
emoid_counts = [89, 1, 91, 2, 13, 6]
bar_colors = ['tab:blue', 'tab:blue', 'tab:red', 'tab:red', 'tab:orange', 'tab:orange']

p = ax[0].bar(cats, rest_counts, color=bar_colors)
ax[0].bar_label(p, label_type='edge')
ax[0].set_title('Resting State Task')
ax[0].set_ylim([0,100])

p = ax[1].bar(cats, nback_counts, color=bar_colors)
ax[1].bar_label(p, label_type='edge')
ax[1].set_title('Working Memory Task')
ax[1].set_ylim([0,100])

p = ax[2].bar(cats, emoid_counts, color=bar_colors)
ax[2].bar_label(p, label_type='edge')
ax[2].set_title('Emotion Identification Task')
ax[2].set_ylim([0,100])

fig.subplots_adjust(top=0.85)

fig.suptitle('Significant (Corrected p<0.1) Correlations Between Tasks/Questionnaire/Demographic Fields\nand FC/VAE Latents in PNC Dataset')
fig.show()

fig.savefig('/home/anton/Documents/Tulane/Research/Work/CL-VAE/pnc-fc-z-corr2.pdf')
```
