```python
import pickle

demofile = '/home/anton/Documents/Tulane/Research/ImageNomer/data/anton/cohorts/UKB/demographics.pkl'

demo = pickle.load(open(demofile, 'rb'))

print(demo.keys())
```

```python
# Read the phenotype info

import pandas as pd

csvname = '/run/media/anton/AntonData1/UKB/Anton_07132023.csv'
df = pd.read_csv(csvname)
df
```

```python
# Load weight and BMI

from math import isnan

bmifield = 'f.21001.0.0'
weightfield = 'f.21002.0.0'

demo[bmifield] = dict()
demo[weightfield] = dict()

for i,sub in enumerate(demo['f_21003_2_0']):
    try:
        bmi = df[df['f.eid'] == sub][bmifield].array[0]
        if not isnan(bmi):
            demo[bmifield][sub] = bmi
    except:
        pass
    try:
        weight = df[df['f.eid'] == sub][weightfield].array[0]
        if not isnan(weight):
            demo[weightfield][sub] = weight
    except:
        pass
    if i % 1000 == 0:
        print(f'Done {i}')
        
print(len(demo[bmifield]))
print(len(demo[weightfield]))
```

```python
# Write new demo

pickle.dump(demo, open(demofile, 'wb'))

print('Complete')
```

```python
# Create B-A scans

import numpy as np

demodir = '/home/anton/Documents/Tulane/Research/ImageNomer/data/anton/cohorts/UKB'

count = 0

for sub in demo['f_21003_3_0']:
    if sub not in demo['f_21003_2_0']:
        continue
    try:
        p0 = np.load(f'{demodir}/fc/{sub}_task-20227-2_fc.npy')
        p1 = np.load(f'{demodir}/fc/{sub}_task-20227-3_fc.npy')
        pdiff = p1-p0
        np.save(f'{demodir}/fc/{sub}_task-20227-3-2-diff_fc.npy', pdiff)
        count += 1
    except:
        continue
        
print(count)
```
