```python
# Find subjects we've preprocessed

predir = '../UKBB_Cypress/func'
presubs = []

from pathlib import Path
import re

exp = re.compile('sub-(\d+)ses(\d)_task-rest_.*\.nii\.gz')

for f in Path(predir).iterdir():
    if not f.is_dir():
        mobj = exp.match(f.name)
        if mobj:
            presubs.append(mobj.groups())
            
print(len(presubs))
```

```python
# Find all subjects with rest ICA

field = 25750
fielddir = f'../UKB_MRI_data_11092022/UKB_MRI/{field}'

presubs = []

exp = re.compile(f'(\d+)_{field}_(\d)_0.txt')

for f in Path(fielddir).iterdir():
    if not f.is_dir():
        mobj = exp.match(f.name)
        if mobj:
            presubs.append(mobj.groups())
            
print(len(presubs))

# presubs = presubs[:10000]
```

```python
# Get 25 comp rfMRI ICA vectorized matrices for the preprocessed subjects

field = 25753
fielddir = f'../UKB_MRI_data_11092022/UKB_MRI/{field}'

import numpy as np

fc25 = dict()
ids2 = []

# 2nd visit

for sub in presubs:
    fname = f'{fielddir}/{sub[0]}_{field}_2_0.txt'
    if Path(fname).exists():
        ids2.append(sub[0])
        
ids2 = sorted(list(set(ids2)))

for sub in ids2:
    fname = f'{fielddir}/{sub}_{field}_2_0.txt'
    data = np.loadtxt(fname)
    fc25[sub] = data

print(len(list(fc25.keys())))
    
```

```python
# Get ages at "2nd" visit, only for those subjects who have fc25

ages2 = [df[df['f.eid'] == int(sub)]['f.21003.2.0'] for sub in ids2]
ages2 = [float(series) for series in ages2]

print(len(ages2))
```

```python
for i in range(10):
    print(ids2[i], ages2[i])
```

```python
import torch
import torch.nn as nn

x = [fc25[sub] for sub in ids2[:4000]]
x = np.stack(x)
x = np.concatenate([x, np.ones((x.shape[0], 1))], axis=1)
y = ages2[:4000]
y = np.array(y)

ntrain = 3500
idcs = torch.randperm(x.shape[0])

xx = torch.from_numpy(x).float().cuda()
xx = xx[idcs]
xtr = xx[:ntrain]
xt = xx[ntrain:]

yy = torch.from_numpy(y).float().cuda()
yy = yy [idcs]
ytr = yy[:ntrain]
yt = yy[ntrain:]

mseLoss = nn.MSELoss()

# w, _, _, _ = torch.linalg.lstsq(xtr.T@xtr+0*torch.eye(xtr.shape[-1]).float().cuda(),xtr.T@ytr)
w, _, _, _ = torch.linalg.lstsq(xtr,ytr)


yhat = torch.mean(ytr)
print(mseLoss(yhat*torch.ones(yt.shape).float().cuda(), yt)**0.5)

yhat = xt@w
print(mseLoss(yhat, yt)**0.5)

```

```python
import matplotlib.pyplot as plt

ytnp = yt.detach().cpu().numpy()
idcs = np.argsort(ytnp)
yhatnp = yhat.detach().cpu().numpy()

print(np.corrcoef(ytnp, yhatnp))

plt.plot(ytnp[idcs])
plt.plot(yhatnp[idcs])
plt.show()
```
