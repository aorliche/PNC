```python
import pickle

demofile = '/home/anton/Documents/Tulane/Research/ImageNomer/data/anton/cohorts/UKB/demographics.pkl'

demo = pickle.load(open(demofile, 'rb'))

print(demo.keys())
```

```python
import pandas as pd

prsfile = '/home/anton/Documents/Tulane/Research/Work/UKB/ukb669358.csv'

df = pd.read_csv(prsfile)

df
```

```python
from math import isnan

fieldsfile = '/home/anton/Documents/Tulane/Research/Work/UKB/fields.ukb'
fields = []

with open(fieldsfile, 'r') as f:
    for line in f:
        fields.append(f'{int(line)}-0.0')
        
for j,field in enumerate(fields):
    demo[field] = dict()
    nancount = 0
    for i,sub in enumerate(demo['f_21003_2_0']):
        try:
            val = df[df['eid'] == sub][field].array[0]
            if isnan(val):
                nancount += 1
                continue
            demo[field][sub] = val
        except Exception as e:
            pass
        if i % 10000 == 0:
            print(f'Done {i}')
    print(f'{j} {len(demo[field])} {nancount}')

print('Complete')
```

```python
# Write new demo

pickle.dump(demo, open(demofile, 'wb'))

print('Complete')
```
