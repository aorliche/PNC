```python
# Read the phenotype info

import pandas as pd

csvname = '/run/media/anton/AntonData1/UKB_MRI_data_11092022/UKB_Phenotypes/project3_Anton/DataSet_11162022.csv'
df = pd.read_csv(csvname)
df
```

```python
import pickle
from pathlib import Path
import re

basedir = '/home/anton/Documents/Tulane/Research/ImageNomer/data/anton/cohorts/UKB/'

demo = pickle.load(open(f'{basedir}/demographics.pkl', 'rb'))
demo['f_21003_2_0'] = dict()
nosex = 0

for i,fc in enumerate(Path(f'{basedir}/fc').iterdir()):
    match = re.match('(\d+)_task-(\d+)-(\d)_fc.npy', fc.name)
    if not match:
        continue
    sub = int(match.group(1))
    field = int(match.group(2))
    scan = int(match.group(3))
    if scan == 2:
        demo['f_21003_2_0'][sub] = float(df[df['f.eid'] == sub]['f.21003.2.0'])
        try:
            demo['f_22001_0_0'][sub] = int(df[df['f.eid'] == sub]['f.22001.0.0'])
        except:
            nosex += 1
    if i % 1000 == 0:
        print(f'Done {i}')
            
print(nosex)
print([len(demo[f]) for f in ['f_21003_2_0', 'f_21003_3_0', 'f_22001_0_0']])
```
