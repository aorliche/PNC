```python
# Make 10-fold CV splits for subjects with both emoid and nback scans and WRAT data

import pickle

# Load meta dict

with open('../../PNC/MegaMeta.bin', 'rb') as f:
    meta = pickle.load(f)
    
print(len(list(meta.keys())))
```

```python
goodIds = []

for key in meta:
    if 'emoid' in meta[key] and 'badEmoid' not in meta[key] and 'nback' in meta[key] and 'wrat' in meta[key]['meta']:
        goodIds.append(key)
        
print(len(goodIds))
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=10, shuffle=True)
groups = []
for split in kf.split(goodIds):
    groups.append(split)
    
print(len(groups))
print(len(groups[0]))
print(len(groups[0][0]))

print(groups[0][1])
print(groups[1][1])
```

```python
import numpy as np

goodIds = np.array(goodIds)

print(goodIds[groups[0][0]])
print(goodIds[groups[0][1]])
```

```python
splits = []

for group in groups:
    splits.append([goodIds[group[0]], goodIds[group[1]]])
    
print(len(splits))
print(len(splits[0]))
print(len(splits[0][0]))
print(len(splits[0][1]))
print(len(splits[1][0]))
```

```python
# Check we don't have repeating subjects in first split

first = splits[0][1]
for split in splits[1:]:
    if np.sum(np.in1d(first, split[1])) > 0:
        print('Contains')
```

```python
# Save splits

with open('../../Work/Explainer/Splits/EmoidNbackWrat10FoldCV.pkl', 'wb') as f:
    pickle.dump(splits, f)
    
print('Complete')
```
