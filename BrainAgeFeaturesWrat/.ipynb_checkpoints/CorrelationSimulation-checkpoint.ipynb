```python
# Import subjects

import pickle

para = 'Nback'

# Load meta dict

with open('../../AllSubjectsMeta.bin', 'rb') as f:
    meta = pickle.load(f)
    
# Load rest subject ids and splits

with open('../../../BrainAgePredictWrat/' + para + 'Only10FoldSplit.bin', 'rb') as f:
    splits = pickle.load(f)
    subids = splits['cand' + para + 'YesWrat']
    groups = splits['groups']
    
print(len(subids))
```

```python
import numpy as np

# Get feature vectors (including true ages)

# Create feature vectors (right now just ages, maleness, and femaless)

X_all = []
for subid in subids:
    subj = meta[subid]
    maleness = 1 if subj['meta']['Gender'] == 'M' else 0
    femaleness = 1 if maleness == 0 else 0
    feat = np.array([subj['meta']['AgeInMonths'], maleness, femaleness])
    X_all.append(feat)
X_all = np.vstack(X_all)

print(X_all[10:20])
print('Complete')
```

```python
# Load WRAT for subjects

import numpy as np

idWratDict = {}

with open('../../wrat.csv') as f:
    for line in f.readlines()[1:]:
        parts = line.split(',')
        try:
            idWratDict[int(parts[0])] = int(parts[3])
        except Exception as e:
            print(parts[0])
            print(e)
            
wrat = np.array([idWratDict[subid] for subid in subids])

print(wrat.shape)
print('Complete')
```

```python
# Get all ages, true and predicted

para = 'Nback'

agesTrue = X_all[:,0]

with open('../../../BrainAgePredictWrat/LeastSquares' + para + 'Predictions.bin', 'rb') as f:
    agesPred = pickle.load(f)
    
with open('../../../BrainAgePredictWrat/Wrat/LeastSquares5' + para + 'WratPredictions.bin', 'rb') as f:
    agesPred5 = pickle.load(f)
    
with open('../../../BrainAgePredictWrat/CNN' + para + 'Predictions.bin', 'rb') as f:
    agesPredCNN = pickle.load(f)
    
with open('../../../BrainAgePredictWrat/MLP' + para + 'Predictions.bin', 'rb') as f:
    agesPredMLP = pickle.load(f)
    
with open('../../../BrainAgePredictWrat/Wrat/TransformerWrat' + para + 'Predictions.bin', 'rb') as f:
    agesPredTrans = pickle.load(f)
    
print('Complete')
```

```python
res = np.stack([X_all[:,1] == 1])
idcs = np.all(res, axis=0)

ap = agesPredTrans[idcs]
at = agesTrue[idcs]
w = wrat[idcs]

corrAB = np.corrcoef(w, at)[0,1]
corrAC = np.corrcoef(w, ap)[0,1]
corrACB = np.corrcoef(w, ap - at)[0,1]

sA = np.cov(w)
sB = np.cov(at)
sC = np.cov(ap)
sBC = np.cov(at, ap)[0,1]

sAC = np.cov(w, ap)[0,1]
sAB = np.cov(w, at)[0,1]
test = (sAC - sAB)/((sA*(sC+sB-2*sBC))**0.5)

print(corrACB)
print(test)
```

```python
# Sliding window graph

from scipy.stats import t

pred = agesPredTrans

def getSigCorr(n,m):
    tv = t.ppf([0.025/m], (n-2))
    b = tv**2/(n-2)
    r = (b/(1+b))**0.5
    return r

winDelta = 12 # +/- months
winStep = 6 # months
winStart = 8*12
winStop = 22*12

winData = np.zeros([3,round((winStop-winStart)/winStep)+1])
winSig = np.zeros(winData.shape)

for g in range(3):
    for i in range(winData.shape[1]):

        if g == 0:
            res = np.stack([X_all[:,1] > -1])
        elif g == 1:
            res = np.stack([X_all[:,1] == 1])
        else:
            res = np.stack([X_all[:,1] == 0])
            
        idcs = np.all(res, axis=0)

        ap = pred[idcs]
        at = agesTrue[idcs]
        w = wrat[idcs]

        winMid = winStart + i*winStep
        winLow = winMid - winDelta
        winHigh = winMid + winDelta

        atIdcs = np.all(np.stack([at > winLow, at < winHigh]), axis=0)

        app = ap[atIdcs]
        att = at[atIdcs]
        ww = w[atIdcs]

        corr = np.corrcoef(app, ww)[0,1]
        winData[g,i] = corr
        winSig[g,i] = getSigCorr(ww.shape[0],winData.size)
#         winData[g,i] = np.mean(app-att)/12
    
print('Complete')
```

```python
import matplotlib.pyplot as plt

ages = np.arange(winData.shape[1])*winStep/12+winStart/12

# Plot significance
plt.fill_between(ages, winSig[0,:], -winSig[0,:], color='#fcc')
plt.plot([8,22], [0,0], color='grey')

plt.plot(ages, winData[0], label='All Subjects')
plt.plot(ages, winData[1], label='Male')
plt.plot(ages, winData[2], label='Female')

plt.title('Emotion Identification Paradigm\nTransformer')

plt.xlabel('Age (Years)')
plt.ylabel(r'Correlation Coefficient $\rho$')

plt.xlim(8,22)
plt.ylim(-1,1)

# plt.legend()
plt.grid()

plt.show()
```
