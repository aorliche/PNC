```python
import pickle

# Load meta dict

with open('../../AllSubjectsMeta.bin', 'rb') as f:
    meta = pickle.load(f)
    
# Load rest subject ids and splits

with open('../../../BrainAgePredictWrat/EmoidOnly10FoldSplit.bin', 'rb') as f:
    splits = pickle.load(f)
    subids = splits['candEmoidYesWrat']
    groups = splits['groups']
    
print(len(subids))
```

```python
# Load wrat

idWratDict = {}

with open('../../wrat.csv') as f:
    for line in f.readlines()[1:]:
        parts = line.split(',')
        try:
            idWratDict[int(parts[0])] = int(parts[3])
        except Exception as e:
            print(parts[0])
            print(e)

print('Complete')
```

```python
import numpy as np

subidsNp = np.array(subids)

# Load timeseries

def loadSeries(prefix, para, idx):
    with open('{:}/{:}_fmri_power264/timeseries/{:}.bin'.format(prefix, para, idx), 'rb') as f:
        return pickle.load(f)

all_ts = [loadSeries('../..', 'emoid', meta[subid]['emoid']) for subid in subidsNp]

print('Loading complete')
```

```python
# Normalize timeseries

def normalizeSubjects(subjects):
    for i in range(len(subjects)):
        subj = subjects[i]
        subj -= np.mean(subj, axis=1, keepdims=True)@np.ones([1,subj.shape[1]])
        subj /= np.std(subj, axis=1, keepdims=True)@np.ones([1,subj.shape[1]])
        if np.sum(np.isnan(subj)) > 0:
            print('nan {:}'.format(i))
        if np.sum(np.isinf(subj)) > 0:
            print('inf {:}'.format(i))
            
normalizeSubjects(all_ts)

print('Complete')
```

```python
# Create feature vectors (right now just ages, maleness, and femaless)

X_all = []
for subid in subidsNp:
    subj = meta[subid]
    maleness = 1 if subj['meta']['Gender'] == 'M' else 0
    femaleness = 1 if maleness == 0 else 0
    feat = np.array([subj['meta']['AgeInMonths'], maleness, femaleness])
    X_all.append(feat)
X_all = np.vstack(X_all)

print(X_all[10:20])
print('Complete')
```

```python
# Calculate pearson matrices

all_p = np.stack([np.corrcoef(ts) for ts in all_ts])

print(all_p.shape)
```

```python
# First order pearson

agesPred = np.zeros([all_p.shape[0]])

# Loop over splits
i = 0
for trainIdcs, testIdcs in groups:
    train_ps = all_p[trainIdcs].reshape(len(trainIdcs), 264*264)
    test_ps = all_p[testIdcs].reshape(len(testIdcs), 264*264)
    
    train_ages = X_all[trainIdcs,0]
    
    w, _, _, _ = np.linalg.lstsq(train_ps, train_ages, rcond=None)
    
    agesPred[testIdcs] = test_ps@w
    
    print('Completed split {:}'.format(i))
    i += 1
    
print('All complete')
```

```python
import matplotlib.pyplot as plt

test_labels = X_all[:,0]
predY = agesPred

idcs = np.argsort(test_labels[:])
yPred = predY[idcs]
y = test_labels[idcs]

corr = np.corrcoef(y, yPred)
rmse = (np.sum((y-yPred)**2)/len(y))**0.5/12
mae = np.sum(np.abs(y-yPred))/12/len(y)

print(corr)
print(rmse)
print(mae)

plt.plot(y, label='Truth')
plt.plot(yPred, label='Prediction')
plt.show()

plt.scatter(y, yPred)
plt.show()
```

```python
# Calculate 5th degree poly

all_5p = np.zeros([len(subids), 264*264*5])
all_5p[:,0:264*264] = all_p.reshape(len(subids), 264*264)

for i in range(1,5):
    all_5p[:,i*264*264:(i+1)*264*264] = all_5p[:,(i-1)*264*264:i*264*264]*all_5p[:,0:264*264]
    
print(all_5p.shape)
```

```python
# First order pearson

agesPred5 = np.zeros([all_5p.shape[0]])

# Loop over splits
i = 0
for trainIdcs, testIdcs in groups:
    train_ps = all_5p[trainIdcs].reshape(len(trainIdcs), 264*264*5)
    test_ps = all_5p[testIdcs].reshape(len(testIdcs), 264*264*5)
    
    train_ages = X_all[trainIdcs,0]
    
    w, _, _, _ = np.linalg.lstsq(train_ps, train_ages, rcond=None)
    
    agesPred5[testIdcs] = test_ps@w
    
    print('Completed split {:}'.format(i))
    i += 1
    
print('All complete')
```

```python
test_labels = X_all[:,0]
predY = agesPred5

idcs = np.argsort(test_labels[:])
yPred = predY[idcs]
y = test_labels[idcs]

corr = np.corrcoef(y, yPred)
rmse = (np.sum((y-yPred)**2)/len(y))**0.5/12
mae = np.sum(np.abs(y-yPred))/12/len(y)

print(corr)
print(rmse)
print(mae)

plt.plot(y, label='Truth')
plt.plot(yPred, label='Prediction')
plt.show()

plt.scatter(y, yPred)
plt.show()
```

```python
# CNN model

import torch
import torch.nn as nn
import torch.nn.functional as F

nc = 10
nHid = 200
nRoi = all_ts[0].shape[0]
nt = all_ts[0].shape[1]

lu = nn.ReLU()
bn = nn.BatchNorm2d(nc).cuda()
dp = nn.Dropout(p=0.05)

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1,nc,(nRoi,5)).cuda()
        self.conv2 = nn.Conv2d(nc,nc,(1,5)).cuda()
        self.fc1 = nn.Linear(nc*(nt-8),nHid).cuda()
        self.fc2 = nn.Linear(nHid,1).cuda()
        
    def forward(self, x):
        nB = x.shape[0]
        x = bn(lu(self.conv1(x.unsqueeze(1))))
        x = lu(self.conv2(x))
        x = x.reshape(nB,-1)
        x = lu(self.fc1(x))
        x = self.fc2(x)
        return x
    
all_ts_t = torch.from_numpy(np.stack(all_ts)).float()
agesPredCNN = torch.zeros(all_ts_t.shape[0])

import random
    
# Loop over splits 
splitIdx = 0
for trainIdcs, testIdcs in groups:
    cnn = CNN()
    optim = torch.optim.Adam(cnn.parameters(), lr=1e-3)
    
    train_ts = all_ts_t[trainIdcs].cuda()
    test_ts = all_ts_t[testIdcs].cuda()
    
    X_train = X_all[trainIdcs,0]
    
    running = 0
    nEpoch = 3000
    pPeriod = 200
    nB = 10
    N = train_ts.shape[0]

    for epoch in range(nEpoch):
        batch = []
        truth = torch.zeros(nB).float().cuda()
        for i in range(nB):
            idx = random.randint(0,N-1)
            batch.append(train_ts[idx])
            truth[i] = X_train[idx]
        batch = torch.stack(batch)
        optim.zero_grad()
        pred = cnn(batch).flatten()
        loss = torch.sum((pred-truth)**2)
        loss.backward()
        running += loss.detach().cpu()
        optim.step()
        if epoch % pPeriod == 0 or epoch == nEpoch-1:
            if epoch != 0:
                if epoch % pPeriod != 0:
                    running /= epoch % pPeriod
                else:
                    running /= pPeriod
            print('epoch {:d} loss={:f}'.format(epoch, ((running/nB)**0.5)/12))
            running = 0

    print('Finished training')
    
    agesPredCNN[testIdcs] = cnn(test_ts).flatten().detach().cpu()
    
    print('Completed split {:}'.format(splitIdx))
    splitIdx += 1

print('All complete')
```

```python
test_labels = X_all[:,0]
predY = agesPredCNN.detach().cpu().numpy()

idcs = np.argsort(test_labels[:])
yPred = predY[idcs]
y = test_labels[idcs]

corr = np.corrcoef(y, yPred)
rmse = (np.sum((y-yPred)**2)/len(y))**0.5/12
mae = np.sum(np.abs(y-yPred))/12/len(y)

print(corr)
print(rmse)
print(mae)

plt.plot(y, label='Truth')
plt.plot(yPred, label='Prediction')
plt.show()

plt.scatter(y, yPred)
plt.show()
```

```python
# MLP

nHid = 80
nFC = 264*264
dp = nn.Dropout(p=0.05)

class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(nFC, nHid).cuda()
        self.fc2 = nn.Linear(nHid, nHid).cuda()
        self.fc3 = nn.Linear(nHid, 1).cuda()
        
    def forward(self, x):
        x = lu(self.fc1(dp(x)))
        x = lu(self.fc2(x))
        return self.fc3(x)
    
# Pearson matrices
all_p_t = torch.from_numpy(all_p).reshape(-1,nFC).float()
agesPredMLP = torch.zeros(all_ts_t.shape[0])

import random
    
# Loop over splits 
splitIdx = 0
for trainIdcs, testIdcs in groups:
    mlp = MLP()
    optim = torch.optim.Adam(mlp.parameters(), lr=1e-3)
    
    train_p = all_p_t[trainIdcs].cuda()
    test_p = all_p_t[testIdcs].cuda()
    
    X_train = X_all[trainIdcs,0]
    
    running = 0
    nEpoch = 3000
    pPeriod = 200
    nB = 10
    N = train_p.shape[0]

    for epoch in range(nEpoch):
        batch = []
        truth = torch.zeros(nB).float().cuda()
        for i in range(nB):
            idx = random.randint(0,N-1)
            batch.append(train_p[idx])
            truth[i] = X_train[idx]
        batch = torch.stack(batch)
        optim.zero_grad()
        pred = mlp(batch).flatten()
        loss = torch.sum((pred-truth)**2)
        loss.backward()
        running += loss.detach().cpu()
        optim.step()
        if epoch % pPeriod == 0 or epoch == nEpoch-1:
            if epoch != 0:
                if epoch % pPeriod != 0:
                    running /= epoch % pPeriod
                else:
                    running /= pPeriod
            print('epoch {:d} loss={:f}'.format(epoch, ((running/nB)**0.5)/12))
            running = 0

    print('Finished training')
    
    agesPredMLP[testIdcs] = mlp(test_p).flatten().detach().cpu()
    
    print('Completed split {:}'.format(splitIdx))
    splitIdx += 1

print('All complete')
```

```python
test_labels = X_all[:,0]
predY = agesPredMLP.detach().cpu().numpy()

idcs = np.argsort(test_labels[:])
yPred = predY[idcs]
y = test_labels[idcs]

corr = np.corrcoef(y, yPred)
rmse = (np.sum((y-yPred)**2)/len(y))**0.5/12
mae = np.sum(np.abs(y-yPred))/12/len(y)

print(corr)
print(rmse)
print(mae)

plt.plot(y, label='Truth')
plt.plot(yPred, label='Prediction')
plt.show()

plt.scatter(y, yPred)
plt.show()
```

```python
# Permuted CNN model

import torch
import torch.nn as nn
import torch.nn.functional as F

nc = 10
nHid = 200
nRoi = all_ts[0].shape[0]
nt = all_ts[0].shape[1]

lu = nn.ReLU()
bn = nn.BatchNorm2d(nc).cuda()
dp = nn.Dropout(p=0.05)

class PCNN(nn.Module):
    def __init__(self):
        super(PCNN, self).__init__()
        self.conv1 = nn.Conv2d(1,nc,(nRoi,5)).cuda()
        self.conv2 = nn.Conv2d(nc,nc,(1,5)).cuda()
        self.fc1 = nn.Linear(nc*(nt-8),nHid).cuda()
        self.fc2 = nn.Linear(nHid,1).cuda()
        
    def forward(self, x):
        nB = x.shape[0]
        x = bn(lu(self.conv1(x.unsqueeze(1))))
        x = lu(self.conv2(x))
        x = x.reshape(nB,-1)
        x = lu(self.fc1(x))
        x = self.fc2(x)
        return x
    
def permute(all_ts):
    rng = np.random.default_rng()
    return rng.permutation(all_ts, axis=2)

all_ts_t = torch.from_numpy(permute(np.stack(all_ts))).float()
print('Finished permutation')

agesPredPCNN = torch.zeros(all_ts_t.shape[0])

import random
    
# Loop over splits 
splitIdx = 0
for trainIdcs, testIdcs in groups:
    cnn = PCNN()
    optim = torch.optim.Adam(cnn.parameters(), lr=1e-3)
    
    train_ts = all_ts_t[trainIdcs].cuda()
    test_ts = all_ts_t[testIdcs].cuda()
    
    X_train = X_all[trainIdcs,0]
    
    running = 0
    nEpoch = 3000
    pPeriod = 200
    nB = 10
    N = train_ts.shape[0]

    for epoch in range(nEpoch):
        batch = []
        truth = torch.zeros(nB).float().cuda()
        for i in range(nB):
            idx = random.randint(0,N-1)
            batch.append(train_ts[idx])
            truth[i] = X_train[idx]
        batch = torch.stack(batch)
        optim.zero_grad()
        pred = cnn(batch).flatten()
        loss = torch.sum((pred-truth)**2)
        loss.backward()
        running += loss.detach().cpu()
        optim.step()
        if epoch % pPeriod == 0 or epoch == nEpoch-1:
            if epoch != 0:
                if epoch % pPeriod != 0:
                    running /= epoch % pPeriod
                else:
                    running /= pPeriod
            print('epoch {:d} loss={:f}'.format(epoch, ((running/nB)**0.5)/12))
            running = 0

    print('Finished training')
    
    agesPredPCNN[testIdcs] = cnn(test_ts).flatten().detach().cpu()
    
    print('Completed split {:}'.format(splitIdx))
    splitIdx += 1

print('All complete')
```
